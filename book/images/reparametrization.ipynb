```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

import mplhep as hep
import hist

import matplotlib.patheffects as path_effects
```

```python
npoints = 200

central_points_r = np.abs(np.random.normal(size=npoints))
central_points_theta = (2*np.random.rand(npoints) - 1) * np.pi

outer_points_r = np.abs(np.random.normal(4, size=2*npoints))
outer_points_theta = (2*np.random.rand(2 * npoints) - 1) * np.pi

fig = plt.figure(figsize=(3,5.5))

ax0 = fig.add_subplot(211)

p1 = ax0.scatter(central_points_r * np.cos(central_points_theta), central_points_r * np.sin(central_points_theta), c='tab:red', marker='.', edgecolor='k', lw=0.25, s=45)
p2 = ax0.scatter(outer_points_r * np.cos(outer_points_theta), outer_points_r * np.sin(outer_points_theta), c='tab:blue', marker='.', edgecolor='k', lw=0.25, s=45)
separation = mpl.patches.Circle((0, 0), 2.25, facecolor='none', edgecolor='k', ls=':', lw=1.25, path_effects=[path_effects.Stroke(linewidth=3, foreground='w'), path_effects.Normal()])
ax0.add_patch(separation)
ax0.set_xlabel('$x$')
ax0.set_ylabel('$y$')
ax0.axis('scaled')
ax0.set_xlim(-6,6)
ax0.set_ylim(-6,6)
ax0.set_xticks([-5, 0, 5])
ax0.set_yticks([-5, 0, 5])

ax2 = fig.add_subplot(212)

ax2.scatter(central_points_r, central_points_theta, c='tab:red', s=45, marker='.', edgecolor='k', lw=0.25)
ax2.scatter(outer_points_r, outer_points_theta, c='tab:blue', s=45, marker='.', edgecolor='k', lw=0.25)
ax2.axvline(2.25, ls=':',c='k', lw=1.25, path_effects=[path_effects.Stroke(linewidth=4, foreground='w'), path_effects.Normal()])
ax2.set_xlabel('$r$')
ax2.set_ylabel('$\\theta$')
ax2.axis('scaled')
ax2.set_xlim(-1,7)
ax2.set_ylim(-4,4)
ax2.set_yticks([-np.pi, 0, np.pi])
ax2.set_yticklabels(['$-\pi$', '0', '$+\pi$'])

'''
ax1 = fig.add_subplot(312, sharex = ax2)

h_in = hist.Hist.new.Regular(25, 0, 6, name='$r$').Double().fill(central_points_r, weight=0.1)
h_out = hist.Hist.new.Regular(25, 0, 6, name='$r$').Double().fill(outer_points_r, weight=0.1)
# ax1.axis('scaled')

hep.histplot(h_in, yerr=False, ax=ax1, color='w', lw=3)
hep.histplot(h_in, yerr=False, ax=ax1, color='tab:red')

hep.histplot(h_out, yerr=False, ax=ax1, color='w', lw=3)
hep.histplot(h_out, yerr=False, ax=ax1, color='tab:blue')
'''

fig.tight_layout()

ax0.legend([p1, p2], ['Data $A$', 'Data $B$'], scatterpoints=3, ncol=2, loc='upper center', bbox_to_anchor=(0.5, +1.25), mode='expand', frameon=False)

fig.savefig('reparametrization.png',bbox_inches='tight',dpi=300)
```
