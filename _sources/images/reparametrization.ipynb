```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
npoints = 100

central_points_r = np.abs(np.random.normal(size=npoints))
central_points_theta = (2*np.random.rand(npoints) - 1) * np.pi

outer_points_r = np.abs(np.random.normal(4, size=2*npoints))
outer_points_theta = (2*np.random.rand(2 * npoints) - 1) * np.pi
fig, axs = plt.subplots(2, 1, figsize=(3,5))
axs[1].scatter(central_points_r, central_points_theta, c='tab:red', s=50, marker='.', edgecolor='k', lw=0.5)
axs[1].scatter(outer_points_r, outer_points_theta, c='tab:blue', s=50, marker='.', edgecolor='k', lw=0.5)
axs[1].axvline(2.25, ls=':',c='k', lw=1)
axs[1].set_xlabel('$r$')
axs[1].set_ylabel('$\\theta$')
axs[1].axis('scaled')
axs[1].set_xlim(-1,7)
axs[1].set_ylim(-4,4)
axs[1].set_yticks([-np.pi, 0, np.pi])
axs[1].set_yticklabels(['$-\pi$', '0', '$+\pi$'])

p1 = axs[0].scatter(central_points_r * np.cos(central_points_theta), central_points_r * np.sin(central_points_theta), c='tab:red', marker='.', edgecolor='k', lw=0.5, s=50)
p2 = axs[0].scatter(outer_points_r * np.cos(outer_points_theta), outer_points_r * np.sin(outer_points_theta), c='tab:blue', marker='.', edgecolor='k', lw=0.5, s=50)
separation = mpl.patches.Circle((0, 0), 2.25, facecolor='none', edgecolor='k', ls=':', lw=1)
axs[0].add_patch(separation)
axs[0].set_xlabel('$x$')
axs[0].set_ylabel('$y$')
axs[0].axis('scaled')
axs[0].set_xlim(-6,6)
axs[0].set_ylim(-6,6)
axs[0].set_xticks([-5, 0, 5])
axs[0].set_yticks([-5, 0, 5])

fig.tight_layout()

axs[0].legend([p1, p2], ['Data $A$', 'Data $B$'], frameon=False, scatterpoints=3, ncol=2, loc='upper center', bbox_to_anchor=(0.5, +1.25))

fig.savefig('reparametrization.png',bbox_inches='tight',dpi=300)

```
