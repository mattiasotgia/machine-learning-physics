```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
npoints = 100

central_points_r = np.abs(np.random.normal(size=npoints))
central_points_theta = (2*np.random.rand(npoints) - 1) * np.pi

outer_points_r = np.abs(np.random.normal(4, size=2*npoints))
outer_points_theta = (2*np.random.rand(2 * npoints) - 1) * np.pi
fig, axs = plt.subplots(2, 1, figsize=(3,5))
axs[1].plot(central_points_r, central_points_theta, '.', c='tab:red')
axs[1].plot(outer_points_r, outer_points_theta, '.', c='tab:blue')
axs[1].axvline(2.25, ls=':',c='k', lw=1)
axs[1].set_xlabel('$r$')
axs[1].set_ylabel('$\\theta$')
axs[1].axis('equal')

axs[0].plot(central_points_r * np.cos(central_points_theta), central_points_r * np.sin(central_points_theta), '.', c='tab:red')
axs[0].plot(outer_points_r * np.cos(outer_points_theta), outer_points_r * np.sin(outer_points_theta), '.', c='tab:blue')
separation = mpl.patches.Circle((0, 0), 2.25, facecolor='none',edgecolor='k', ls=':', lw=1)
axs[0].add_patch(separation)
axs[0].set_xlabel('$x$')
axs[0].set_ylabel('$y$')
axs[0].axis('equal')

fig.tight_layout()
fig.savefig('reparametrization.png',bbox_inches='tight',dpi=300)
```
